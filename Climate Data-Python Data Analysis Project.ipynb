# Q2) Find the number of times when the " Weather is exactly Clear" ?

```python
data.head(2)    #this are dataset
```

```python
There are 3 ways of solving 

1 value_counts

2 Filtering

3 Groupby
```

```python
#value_counts()

data['Weather'].value_counts()                     ### Total Clear -- is 1326
```

```python
#Filtering

data[data.Weather == 'Clear']               #1326 rows all are clear
```

```python
#Groupby 


data.groupby('Weather').get_group('Clear')
```
